```python
# importando pacotes
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

# configurando randomização
np.random.seed(19680801)
```

```python
# definindo variáveis
X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))
```

```python
# criando canvas
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1, aspect=1)
```

```python
# definindo função
def minor_tick(x, pos):
    if not x % 1.0:
      return ""
    return "%.2f" % x
```

```python
#determina a razão de crescimento do eixo X -> 1 em 1
ax.xaxis.set_major_locator(MultipleLocator(1.000))  
#determina a razão de crescimento do fractal do eixo X, 0.25 em 0.25 irá formar a escala 1 em 1 da linha acima
ax.xaxis.set_minor_locator(AutoMinorLocator(4))  

#determina a razão de crescimento do eixo y -> 1 em 1
ax.yaxis.set_major_locator(MultipleLocator(1.000))
#determina a razão de crescimento do fractal do eixo y, 0.25 em 0.25 irá formar a escala 1 em 1 da linha acima
ax.yaxis.set_minor_locator(AutoMinorLocator(4))

# plota no gráfico os valores referentes a cada eixo
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))
ax.yaxis.set_minor_formatter(FuncFormatter(minor_tick))

# determina o tamanho da matriz do gráfico 
ax.set_xlim(0,5)
ax.set_ylim(0,4)

#plotando fig
fig
```

```python
# configurando layout 'major'
ax.tick_params(which='major', width=5.0)
ax.tick_params(which='major', length=15)

#plotando fig
fig
```

```python
# configurando layout 'minor'
ax.tick_params(which='minor', width=3.0, labelsize=10)
ax.tick_params(which='minor', length=7, labelcolor = '0.45') # quanto maior o numero da cor, mais claro é impresso

#plotando fig
fig
```

```python
# configurando grid
ax.grid(linestyle='--', linewidth=0.9, color='0.75', zorder=-10) # quanto maior o numero da cor, mais claro é impresso

#plotando fig
fig
```

```python
# plotando valores


# plotando linha azul
ax.plot(X, Y1, c=(0.25, 0.25, 1), lw=5, label="Blue signal", zorder=10)
#plotando linha vermelha
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
#plotando valores aleatorios
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

#plotando gráfico
fig
```

```python
# plotando legendas

#titulo
ax.set_title("Anatomy of a Figure", fontsize=20, verticalalignment='bottom')
# legenda eixo X
ax.set_xlabel("X axis Label")
# legenda eixo Y
ax.set_ylabel("Y axis Label")

#plotando grafico
fig
```

```python
# criando funcões de desenho
def circle(x,y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x,y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0,0,0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)

def text(x,y, text):
    ax.text(x,y, text, backgroundcolor='white',
            ha='center', va='top', weight='bold', color='blue')
```

```python
# utilizando funções de desenho criadas
# para marcar pontos de destaque no gráfico

# minor tick label
circle(0.5, -0.1)
text(0.5, -0.32, 'Minor Tick label')

# major tick label
circle(-0.25, 3.00)
text(-0.25, 2.80, 'Major Tick Label')

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# X Label
circle(2.25, -0.35)
text(2.00, -0.6, "X axis label")

# Y Label
circle(-0.5, 1.70)
text(-0.5, 1.50, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

#plotando fig
fig
```

```python
# fazendo anotações
color = 'blue'
ax.annotate('Spines', xy= (4.0,0.35), xycoords='data',
            xytext=(3.3,0.5), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle='arc3',
                            color=color))
ax.annotate('', xy=(3.15, 0.0), xycoords='data',
            xytext=(3.45, 0.45), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

# plotando fig
fig
```
